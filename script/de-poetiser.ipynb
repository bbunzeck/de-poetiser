```python
import nltk
import random
```

```python
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet 
```

```python
f = open('input.txt', 'r')
content = f.read()
print(content)
```

```python
words = content.split()
new_content = list()
for word in words:
    syn = list()
    for synset in wordnet.synsets(word):
        for lemma in synset.lemmas():
            syn.append(lemma.name())
    if len(syn) > 0:
        new_content.append(random.choice(syn))
    else:
        new_content.append(word)
separator = ' '
output = separator.join(new_content)
```
