# De-poetiser

A jupyter notebook that de-poetises texts by replacing words with random synonyms from WordNet, effectively destroying rhymes, rhythm, and changing connotation, metaphors and analogies.

## Prerequisites

NLTK and the included [WordNet Corpus](https://wordnet.princeton.edu/) need to be installed

```python
import nltk
import random
from nltk.corpus import wordnet, stopwords
```

## De-poetisation

The text sample is the first chapter of The Great Gatsby. 

```python
f = open('input.txt', 'r')
content = f.read()
print(content)
```

```python
words = content.split()
new_content = list()
for word in words:
    if word in stopwords.words('english'):
        new_content.append(word)
    else:
        syn = list()
        for synset in wordnet.synsets(word):
            for lemma in synset.lemmas():
                syn.append(lemma.name())
        if len(syn) > 0:
            new_content.append(random.choice(syn))
        else:
            new_content.append(word)
separator = ' '
output = separator.join(new_content)
```
